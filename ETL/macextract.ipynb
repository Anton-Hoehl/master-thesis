```python
import os
import numpy as np 
import pandas as pd

from datetime import datetime
import pickle
import fitz  
from PIL import Image
```

```python

def interpret_date(rawstr):

    splits = rawstr.split('_')
    journal = splits[0]
    date_str = splits[1]

    # Extract the year and month parts
    year_part = int(date_str[:2])
    month_part = int(date_str[2:])
    
    # Determine the full year considering the millennium break
    if year_part >= 50:
        full_year = 1900 + year_part
    else:
        full_year = 2000 + year_part
    
    # Create a datetime object (day is set to 1)
    date_obj = datetime(full_year, month_part, 1)
    
    return [journal, date_obj]
```

```python
# change path if necessary 
path_pdfs = '/home/adm_gpu/projects/masterthesis/code/test_scraping/macworld_pdfs/'
l_pdfs = os.listdir(path_pdfs)
```

```python

l_images = []

for name in l_pdfs:
    infolist = interpret_date(name)
    print(name, infolist)
    fullpath = path_pdfs + name

    for index in range(10):

        pdf_document = fitz.open(fullpath)

        page = pdf_document.load_page(index +1)

        image = page.get_pixmap()

        image_pil = Image.frombytes("RGB", [image.width, image.height], image.samples)
        #print(infolist.append(image_pil))
        l_images.append([infolist[0], infolist[1], image_pil])
```
