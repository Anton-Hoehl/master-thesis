```python
# imports 
import os 
import re
import time
from lxml import html
import numpy as np 
import requests
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# strings

target_folder = 'macworld_pdfs'
link = 'https://vintageapple.org/macworld/'
```

```python
# functions


def download_pdf(url, folder_path):
    # Create the folder if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Extract the filename from the URL
    filename = url.split('/')[-1]

    time.sleep(30)
    print(85*'=')
    print(url)
    # Download the PDF
    response = requests.get(url)
    print(response.status_code)
    if response.status_code == 200:
        # Save the PDF to the specified folder
        with open(os.path.join(folder_path, filename), 'wb') as f:
            f.write(response.content)
        print(f"PDF downloaded and saved as {filename} in {folder_path}")
    else:
        print("Failed to download PDF")


```

```python
# pdf links 


page = requests.get(link)
webpage = html.fromstring(page.content)

full_linklist = webpage.xpath('//a/@href')
```

```python
# filter for pdfs 

reg = re.compile(r'\.pdf')

pdf_list = list(filter(reg.search, full_linklist))

new_reg = re.compile('0\d\d\d')


newlist = list(filter(new_reg.search, pdf_list))
print(newlist)
```

```python


newlist[0]
```

```python
test = link + newlist[0]

print(test)
download_pdf(test, target_folder)
```

```python
for edition in newlist:
    finallink = link + edition
    download_pdf(finallink, target_folder)
```
